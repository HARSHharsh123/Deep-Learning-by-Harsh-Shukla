```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_or = pd.DataFrame()
df_and = pd.DataFrame()
df_xor = pd.DataFrame()
```

```python
df_or['input1'] = [1,1,0,0]
df_or['input2'] = [1,0,1,0]
df_or['ouput'] = [1,1,1,0]
```

```python
df_or
```

```python
plt.rcParams['figure.figsize'] = [7,4]
sns.scatterplot(x = df_or['input1'] , y = df_or['input2'] , hue = df_or['ouput'] , s = 200) 
```

```python
df_and['input1'] = [1,1,0,0]
df_and['input2'] = [1,0,1,0]
df_and['ouput'] = [1,0,0,0]
```

```python
df_and
```

```python
plt.rcParams['figure.figsize'] = [7,4]
sns.scatterplot(x = df_and['input1'] , y = df_and['input2'] , hue = df_and['ouput'] , s = 200) 
```

```python
df_xor['input1'] = [1,1,0,0]
df_xor['input2'] = [1,0,1,0]
df_xor['ouput'] = [0,1,1,0]
```

```python
df_xor
```

```python
plt.rcParams['figure.figsize'] = [7,4]
sns.scatterplot(x = df_xor['input1'] , y = df_xor['input2'] , hue = df_xor['ouput'] , s = 200) 
```

```python
## Traning of Perceptron 

from sklearn.linear_model import Perceptron 

p_or = Perceptron()
p_and = Perceptron()
p_xor = Perceptron()
```

```python
p_or.fit(df_or.iloc[:,:-1] , df_or.iloc[:,-1])
p_and.fit(df_and.iloc[:,:-1] , df_and.iloc[:,-1])
p_xor.fit(df_xor.iloc[:,:-1] , df_xor.iloc[:,-1])
```
