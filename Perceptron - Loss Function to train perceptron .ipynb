```python
from sklearn.datasets import make_classification
X , y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=15)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(x = X[:,0] , y = X[:,1], hue = y)
```

```python
def Perceptron(X , y):
    w1 = w2 = b = 1
    lr = 0.1
    for j in range(1000):
        for i in range(X.shape[0]): ## we are running our loop for 100 data values / points 
            ## finding f(xi) or z 
            z = w1 * X[i][0] + w2 * X[i][1] + b
            ## checking condition 
            if z * y[i] < 0: ## and after updating values of w1 , w2 , b 
                w1 = w1 + lr * y[i] * X[i][0]
                w2 = w2 + lr * y[i] * X[i][1]
                b = b + lr * y[i]
    return w1 , w2 , b
```

```python
w1 , w2 , b = Perceptron(X , y)
```

```python
## finding m and c values through w1 , w2 , b values 

m = -(w1 / w2)
c = -(b / w2)
```
